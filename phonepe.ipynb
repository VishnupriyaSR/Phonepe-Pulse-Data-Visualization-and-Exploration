# Data transformation

```python
# Dataframe of aggregated Transactions

path="C:/Users/User/project2/data/pulse/data/aggregated/transaction/country/india/state/"

states_name_list=os.listdir(path)   #get the list of states in India

column1={'State':[], 'Year':[],'Quarter':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for state in states_name_list:
    cur_state = path + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            A=json.load(data)
            
            for i in A["data"]["transactionData"]:
                Name=i['name']
                Count=i['paymentInstruments'][0]['count']
                Amount=i['paymentInstruments'][0]['amount']
                
                column1["Transacion_type"].append(Name)
                column1["Transacion_count"].append(Count)
                column1["Transacion_amount"].append(Amount)
                column1["State"].append(state)
                column1["Year"].append(year)
                column1["Quarter"].append(int(file.strip(".json")))
Aggre_Trans=pd.DataFrame(column1)
Aggre_Trans                         
```

```python
#Dataframe of Aggregated Users

path1="C:/Users/User/project2/data/pulse/data/aggregated/user/country/india/state/"

states_name_list=os.listdir(path1)

column2={"State":[],"Year":[],"Quarter":[],"Brand":[],"Count":[],"Percentage":[]}

for state in states_name_list:
    cur_state = path1 + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            B=json.load(data)
            try:
                for j in B["data"]["usersByDevice"]:
                    Brand=j["brand"]
                    Count=j["count"]
                    Percentage=j["percentage"]

                    column2["Brand"].append(Brand)
                    column2["Count"].append(Count)
                    column2["Percentage"].append(Percentage) 
                    column2["State"].append(state)
                    column2["Year"].append(year)
                    column2["Quarter"].append(int(file.strip(".json")))
            except:
                pass

Aggre_Users=pd.DataFrame(column2)
Aggre_Users                              
```

```python
#Dataframe of map transactions

path2="C:/Users/User/project2/data/pulse/data/map/transaction/hover/country/india/state/"

states_name_list=os.listdir(path2)

column3={"State":[],"Year":[],"Quarter":[],"District":[],"Count":[],"Amount":[]}

for state in states_name_list:
    cur_state = path2 + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            C=json.load(data)
            
            for j in C["data"]["hoverDataList"]:
                Name=j["name"]
                Count=j["metric"][0]["count"]
                Amount=j["metric"][0]["amount"] 
                column3["District"].append(Name)
                column3["Count"].append(Count)
                column3["Amount"].append(Amount)
                column3["State"].append(state)
                column3["Year"].append(year)
                column3["Quarter"].append(int(file.strip(".json")))

Map_Trans=pd.DataFrame(column3)
Map_Trans                
```

```python
#Dataframe of map Users

path3="C:/Users/User/project2/data/pulse/data/map/user/hover/country/india/state/"

states_name_list=os.listdir(path3)

column4={"State":[],"Year":[],"Quarter":[],"District":[],"RegisteredUsers":[],"AppOpens":[]}

for state in states_name_list:
    cur_state = path3 + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            D=json.load(data)
            
            for i in D["data"]["hoverData"].items():
                Name=i[0]
                RegisteredUser=i[1]["registeredUsers"]
                AppOpens=i[1]["registeredUsers"]
                
                column4["District"].append(Name)
                column4["RegisteredUsers"].append(RegisteredUser)
                column4["AppOpens"].append(AppOpens)
                column4["State"].append(state)
                column4["Year"].append(year)
                column4["Quarter"].append(int(file.strip(".json")))

Map_Users=pd.DataFrame(column4)
Map_Users 
```

```python
#Dataframe of Top transactions

path3="C:/Users/User/project2/data/pulse/data/top/transaction/country/india/state/"

states_name_list=os.listdir(path3)

column5={"State":[],"Year":[],"Quarter":[],"Pincode":[],'Transacion_count':[], 'Transacion_amount':[]}

for state in states_name_list:
    cur_state = path3 + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            E=json.load(data)
            
            for k in E["data"]["pincodes"]:
                Name=k["entityName"]
                Count=k["metric"]["count"]
                Amount=k["metric"]["amount"] 
                column5["Pincode"].append(Name)
                column5["Transacion_count"].append(Count)
                column5["Transacion_amount"].append(Amount)
                column5["State"].append(state)
                column5["Year"].append(year)
                column5["Quarter"].append(int(file.strip(".json")))

Top_Trans=pd.DataFrame(column5)
Top_Trans                
```

```python
#Dataframe of Top Users


path4="C:/Users/User/project2/data/pulse/data/top/user/country/india/state/"

states_name_list=os.listdir(path4)

column6={"State":[],"Year":[],"Quarter":[],"Pincode":[],"RegisteredUsers":[]}


for state in states_name_list:
    cur_state = path4 + state + "/"
    agg_year=os.listdir(cur_state)
    
    for year in agg_year:
        cur_year=cur_state + year + "/"
        agg_year_list=os.listdir(cur_year)
        
        for file in agg_year_list:
            cur_json=cur_year + file
            data=open(cur_json,"r")
            F=json.load(data)
            
            
            for k in F["data"]["pincodes"]:
                Name=k["name"]
                RegisteredUsers=k["registeredUsers"]
                column6["Pincode"].append(Name)
                column6["RegisteredUsers"].append(RegisteredUsers)
                column6["State"].append(state)
                column6["Year"].append(year)
                column6["Quarter"].append(int(file.strip(".json")))

Top_Users=pd.DataFrame(column6)
Top_Users               
```

```python
#Connect to the POSTgre-SQL Server
mydb=psycopg2.connect(host="localhost",user="postgres",password="vishnu",database="Phonepe",port=5432)
cursor=mydb.cursor()

#Creating aggregated-transaction table

create_aggtrans="create table if not exists agg_trans(State varchar(100),Year int,Quarter int,Transaction_type varchar(100),Transaction_count int,Transaction_amount double precision)"
cursor.execute(create_aggtrans)

#insert values into aggregated-transaction table
for i,row in Aggre_Trans.iterrows():
    query="insert into agg_trans values(%s,%s,%s,%s,%s,%s)"
    cursor.execute(query,tuple(row))
    mydb.commit()
```

```python
#Creating aggregated-user table
create_aggusers="create table if not exists agg_users(State varchar(100),Year int,Quarter int,Brand varchar(100),Count int,Percentage double precision)"
cursor.execute(create_aggusers)

#insert values into aggregated-user table
for i,row in Aggre_Users.iterrows():
    query="insert into agg_users values(%s,%s,%s,%s,%s,%s)"
    cursor.execute(query,tuple(row))
    mydb.commit()
```

```python
#Creating Map-transaction table
create_maptrans="create table if not exists map_trans(State varchar(100),Year int,Quarter int,District varchar(100),Count int,Amount double precision)"
cursor.execute(create_maptrans)

#insert values into Map-transaction table
for i,row in Map_Trans.iterrows():
    query="insert into map_trans values(%s,%s,%s,%s,%s,%s)"
    cursor.execute(query,tuple(row))
    mydb.commit()
```

```python
#Creating Map-Users table
create_mapusers="create table if not exists map_users(State varchar(100),Year int,Quarter int,District varchar(100),RegisteredUsers int,AppOpens int)"
cursor.execute(create_mapusers)

#insert values into Map-Users table
for i,row in Map_Users.iterrows():
    query="insert into map_users values(%s,%s,%s,%s,%s,%s)"
    cursor.execute(query,tuple(row))
    mydb.commit()
```

```python
#Creating Top-Transactions table
create_toptrans="create table if not exists top_trans(State varchar(100),Year int,Quarter int,Pincode int,Transaction_count int,Transaction_amount double precision)"
cursor.execute(create_toptrans)

#insert values into Top-Transactions table
for i,row in Top_Trans.iterrows():
    query="insert into top_trans values(%s,%s,%s,%s,%s,%s)"
    cursor.execute(query,tuple(row))
    mydb.commit()
```
